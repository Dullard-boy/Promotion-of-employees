```python
test['is_promoted']='0'
test
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score
feature_cols=['previous_year_rating','KPIs_met >80%','awards_won?','avg_training_score']
x_train=train[feature_cols]
y_train=train.is_promoted
x_test=test[feature_cols]
y_test=test.is_promoted
clf=DecisionTreeClassifier()
clf=clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
pred=pd.DataFrame(y_pred)
pred.value_counts()
```

```python
pred=pred.rename(columns={0:'is_promoted'})
pred
```

```python
test.update(pred, overwrite=True)
test=test.groupby('is_promoted')
print('Employees Promoted')
test.get_group(1)
```

```python
test=test.get_group(1)
print('Statistical Representation of Promoted Employees in each department:')
test['department'].value_counts().plot.bar()
```
